```python
m = 0
b = 0

learning_rate = 0.001
for i in range(len(x)):
    x1 = x.iloc[0,:].values[0]
    y1 = y.iloc[0,:].values[0]
    guess = m * x1 + b
    error = y1 - guess
    m -= error * x1 * learning_rate
    b -= error * learning_rate
    print(i,"-->",error)
```

```python
def gd(x,y,n):
    m=6
    c=3
    alpha=0.8
    len=x.shape[0]
    for r in range(n):
        y_pred=m*x +c
        cost=(1/len)*((y_pred-y)**2).sum()
        print('cost is ',cost)
        m=m-(-alpha/len)*((y-y_pred)*x).sum()
        c=c-(-alpha/len)*((y-y_pred).sum())
        print('m,c is',m,c)
        if cost<1:
            plt.scatter(x,y)
            plt.plot(x,y_pred)
```
