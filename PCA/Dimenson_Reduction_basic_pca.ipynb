```python
import numpy as np
A = np.array([
          [3,7],
          [-4,-6],
          [7,8],
          [1,-1],
          [-4,-1],
          [-3,-7]])
```

```python
A
```

```python
A.shape
```

```python
import pandas as pd
df = pd.DataFrame(A,columns = ['a0','a1'])
df
```

```python
A0= A[:,0]
A1 = A[:, 1]
```

```python
A0
```

```python
A1
```

```python
#There are 3 ways to find covariance
# a) summation (x_i - x_bar)(y_i - y_bar)^2/(n-1)
# b) Trasnpose method and its matrix multiplication
# c) 
np.cov(A0,A1) 
```

```python
np.sum(A0*A1)/5 #n-1
```

```python
#another way to find covariance
A.T
```

```python
sigma   = (A.T @ A)/5 # @ is matrix multiplication
```

```python
sigma
```

```python
l , x = np.linalg.eig(sigma)
```

```python
l #eigen values 56.92
```

```python
x #eigen vectors as l(eigen value) has bigger value of 56.92 this is PC1 at x[:,1]
```

```python
sigma@x[:,0]
```

```python
sigma@x[:,1]
```

```python
x
```

```python
print("1st pc")
x[:,1]
```

```python
print("2nd pc")
x[:,0]
```

```python
pc_1 = A@x[:,1] #transformed as per 1st pc
pc_1
```

```python
pc_2 = A@x[:,0] #transformed as per 2nd pc
pc_2
```

```python
#standard normal distribution to convert to mean = 0 and std = 1
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(df)
```

```python
type(scaled)
```

```python
df_scaled = pd.DataFrame(scaled)
df_scaled
```

```python
df_scaled.describe()
```

```python
n = df_scaled.shape[0]
```

```python
mat_mul = (df_scaled.T @ df_scaled)/(n-1)
mat_mul
```

```python
l , x = np.linalg.eig(mat_mul)
```

```python
l #eigen value 
```

```python
max(l)
```

```python
x #eigen vector #choose first list of vector for PC1 
```

```python
pc1 = df_scaled @ x[:,0]
pc1
```

```python
pc2 = df_scaled @ x[:,1]
pc2
```

```python
checking = np.dot(pc1, pc2) #nearly 0 as pc1 and pc2 are orthogonal
checking
```

```python
#sklearn
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 3)
```

```python
pc1
```

```python
pca.fit_transform(df_scaled) #note sign is reversed in fit_tranform and pc1
```

```python

pca.explained_variance_ratio_
```

```python
pca = PCA()
```
