```python
#importing necessary libraries

import pandas as pd 
import numpy as np 
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn.metrics import accuracy_score , confusion_matrix , roc_auc_score, roc_curve 
import pandas.util.testing as tm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 20) #25% for testing
```

```python
knn1 = KNeighborsClassifier() #5 is default neighbors. use hypertuning to find neighbors
```

```python
knn1.fit(x_train, y_train)
```

```python
knn1.score(x_test,y_test)
```

```python
#tuning iwth n_neighbors
param = {
        'n_neighbors' : [5,10,15,20,25,30,35,40,50,55,60,65,70,100,120]
         }
```

```python
grid_cv = GridSearchCV(knn1, param_grid = param)
```

```python
grid_cv.fit(x_train,y_train)
```

```python
grid_cv.best_params_
```

```python
knn_new2 = KNeighborsClassifier(n_neighbors=65)
```
